```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup("https://www.sharesansar.com/live-trading", 'html.parser')
soup=requests.get(url)
soup
```

```python
soup.text   # To extract the source code inside pc
```

```python
code=BeautifulSoup(soup.text,"lxml") 
code
```

```python
table_code=code.table
table_code
```

```python
tags=table_code.find_all('tr')
tags
```

```python
list_data=[]
for i in tags:
    data=i.text
    list_data.append(data)
list_data
```

```python
cleaned_data=[]
for i in list_data[1:]:
    x=i.replace("                                        ","")
    x=x.split("\n\n")
    x=[j.replace('\n','') for j in x]
    cleaned_data.append(x[1:])

cleaned_data
```

```python
col=list_data[0].split('\n')[1:-1]  # !st line of the list_data
col
```

```python
import pandas as pd
df=pd.DataFrame(cleaned_data,columns=col)  #columns=col----> make columns with all the data of col
df
```

```python
try:
    df['LTP']=[float(i.replace("," , "")) for i in df['LTP']]
    df['Point Change']=[float(i.replace("," , "")) for i in df['Point Change']]
    df['% Change']=[float(i) for i in df['% Change']]
    df['Open']=[float(i.replace("," , "")) for i in df['Open']]
    df['High']=[float(i.replace("," , "")) for i in df['High']]
    df['Low']=[float(i.replace("," , "")) for i in df['Low']]
    df['Volume']=[float(i.replace("," , "")) for i in df['Volume']]
    df['Prev. Close']=[float(i.replace("," , "")) for i in df['Prev. Close']]
except:
    print("Already changed into float")
```
